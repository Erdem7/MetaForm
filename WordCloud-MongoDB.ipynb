```python
from pymongo import MongoClient

import pandas as pd 
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

import numpy as np
from os import path
from PIL import Image
import collections
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
client = MongoClient("mongodb://localhost:27017")
db = client.example
```

```python
collection_conn = db['F35']
collection_cursor = collection_conn.find()
collection_pandas_df = pd.DataFrame(list(collection_cursor))
```

```python
texts=collection_pandas_df.datas[0]
```

```python
print(texts)
```

```python
wordcloud = WordCloud().generate(texts)
```

```python
plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
filtered_words_spam = [word for word in texts.split() if word not in STOPWORDS]
counted_words_spam = collections.Counter(filtered_words_spam)
word_count_spam = {}
for letter, count in counted_words_spam.most_common(30):
    word_count_spam[letter] = count
for i,j in word_count_spam.items():
        print('Word: {0}, count: {1}'.format(i,j))    
```
